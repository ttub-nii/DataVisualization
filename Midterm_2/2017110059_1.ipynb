```python
import csv
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 1. 출생지역의 남여구분없이 다운로드한 데이터를 사용하여,
birth = input('인구 구조가 알고 싶은 지역 이름(동 단위)을 입력해주십시오.  ex.병점2동   ')
file1 = open('2009MaleFemaleNo.csv', 'r', encoding = 'cp949')   # 2009년 자료
file2 = open('2018MaleFemaleNo.csv', 'r', encoding = 'cp949')   # 2018년 자료
data1 = csv.reader(file1)
data2 = csv.reader(file2)
result1 = []
result2 = []

# 2. 출생지(동 단위)를 입력받아
for row in data1:
    if birth in row[0]:
        for i in row[3:]:
            result1.append(int(i))
print(result1)

for row in data2:
    if birth in row[0]:
        for i in row[3:]:
            result2.append(int(i))
print(result2)

# matplotlib에서 사용 가능한 한글 지원 ttf 폰트가 없어 직접 지정한다.
path = '/Applications/anaconda3/lib/python3.7/site-packages/matplotlib/mpl-data/fonts/ttf/NanumSquareRegular.ttf'
fontprop = fm.FontProperties(fname=path, size=13)

# 3. 2009년과 2018년의 각각 해당하는 연령별 인구수를 꺽은선으로 표현합니다.
plt.title(birth +' 지역의 연령별 인구수 그래프', fontproperties=fontprop)
plt.style.use('ggplot')
plt.plot(result1, marker='o', label='2009')
plt.plot(result2, marker='o', label='2018')
plt.savefig('2017110059_1.png', dpi=300)
plt.show()
```
